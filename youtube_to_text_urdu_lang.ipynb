```python
x = "/content/output/janbaz/vocals.wav"

import csv
with open('/content/output/train.tsv', 'w') as out_file:
    for i in range(0,30000,5):
    
    
    newAudio = AudioSegment.from_wav("/content/output/janbaz/vocals.wav")
    t1 = i * 1000
    j = i+5
    t2 = j * 1000
    newAudio = newAudio[t1:t2]
    newAudio.export("pc-"+str(i)+'.wav', format="wav") #Exports to a wav file in the current path.
    r = sr.Recognizer()
  #pas the downloaded wav file name here !!!!!!
    aud = "pc-"+str(i)+".wav"
    voco = aud
    vocals = "pc-"+str(i)+"/"
    voc = !spleeter separate -o output/ {voco}
    with sr.AudioFile("output/"+vocals+"vocals.wav") as source:
        audio =  r.record(source)
        try:
                      
            
            
            text = r.recognize_google(audio,language='ur')
            
            tsv_writer = csv.writer(out_file, delimiter='\t')
            tsv_writer.writerow(['6cf52fee343629ae9a854bce2bc137221c710d167119554ff1f30afecaace31274cbd3293f2f383647c55e51ce7711b9e8842b8283189ddec40394d1414d5dcd', aud,text,"","","","","","ur",""])
            print ("This text extracted from given audio:"+aud+"is : " + text)
        #error occurs when google could not understand what was said
        except sr.UnknownValueError:
            print("Google Speech Recognition could not understand audio")
        except sr.RequestError as e:
            print("Could not request results from Google Speech Recognition service; {0}".format(e))
    
```
